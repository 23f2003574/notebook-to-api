```python
def add(a: int, b: int) -> int:
    return a + b


def multiply(a: int, b: int) -> int:
    return a * b


def predict(age: int, salary: float, city: str):
    return {
        "score": age * salary,
        "message": f"User in {city} processed."
    }


x = add(2, 3)
y = multiply(2, 3)
z = predict(25, 50000.0, "Berlin")
print(x, y, z)
```
